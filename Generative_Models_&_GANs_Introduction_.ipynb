## Intro to Genrative Model

A generative model is a kind of machine learning model that seeks to produce new, comparable data by understanding the underlying patterns or distributions of the original data. It's essentially the same as instructing a computer to create its own data by using what it has already seen. This model is important because it may produce, which has broad applications in a variety of domains, including science and the arts.

Generative models is an essential component of artificial intelligence (AI). Their main job is to analyze and extract the underlying distributions or patterns from a given batch of data. The model can produce new data that is comparable to the original dataset once it has mastered these patterns.

Assume you are instructing a little child in animal drawing. The youngster starts to comprehend the basic characteristics of each animal after being shown multiple images of various animals. With enough time, the child may combine learnt features to create a drawing of an animal they have never seen before. This is similar to the way a generative model works, which is to learn from the data it is exposed to and then generate new content based on that understanding.

In machine learning, the difference between generative and discriminative models is crucial.

**Generative models:** The goal of these models is to comprehend the process of data generation. Their goal is to understand the data's distribution. A generative model would, for example, attempt to determine what characteristics distinguish a dog from a cat in an image of the two animals. After that, it could produce fresh visuals that either looked like cats or dogs.

**Discriminative models:** In contrast, these models concentrate on differentiating between various kinds of data. They learn the boundaries that set one kind of data apart from another, rather not necessarily learning or comprehending how the data is formed. A discriminative model would be able to distinguish between cats and dogs using the same example, but it might not be able to create a new image of a cat or dog on its own.

In the field of artificial intelligence, generative models are essential for tasks involving the production of new content. This might take the shape of creating writing, creating music, or even synthesizing lifelike human faces. Their capacity to "dream up" fresh data makes them indispensable in situations requiring either the augmentation of pre-existing datasets or the creation of original information.

In simpler words, generative models thrive at creation, but discriminative models perform well on classification tasks. Given their creative ability and in-depth knowledge of data distributions, generative models are a potent weapon in the AI toolbox.

## Types of Generative Models

###  Generative Adversarial Networks (GANs)
These models are like artistic rivals. A generator continuously creates new data, and a discriminator tries to determine if the generated data is real or fake. Through this competition, the generator learns to produce increasingly realistic data, while the discriminator hones its ability to detect forgeries.

### Variational Autoencoders (VAEs)

VAEs work in two parts: an encoder and a decoder. The encoder condenses the input data into a compressed form, like a code capturing its essence. The decoder then uses this code to reconstruct the original data or generate similar new data. VAEs are useful for tasks like dimensionality reduction and data compression.

image.png


### Autoregressive models

 These models generate data one piece at a time, like building a sentence word by word. They analyze the patterns in sequences of data and predict the next element based on what has already been generated. Recurrent neural networks (RNNs) and Transformer-based models are popular architectures for autoregressive models.

image.png

### Gaussian Mixture Models (GMMs)

These models assume the data originates from a blend of several Gaussian distributions (bell-shaped curves). By figuring out the properties of these distributions, GMMs can generate new data points that probabilistically fit the mold. They are useful for tasks like clustering and anomaly detection.

image.png

### Diffusion models

Imagine gradually adding noise to an image until it becomes completely random. Diffusion models work in reverse. They begin with random noise and progressively remove it, step by step, to create a realistic image. This approach is particularly well-suited for generating high-fidelity images.

## Introduction to GANs

Generative Adversarial Networks (GANs) were introduced in 2014 by Ian Goodfellow and his team. GANs are a generative modeling technique that creates new data resembling the training data. They consist of two primary components—two neural networks—that compete against each other, enabling them to capture, replicate, and analyze the variations within a dataset. These two models are typically referred to as the Generator and the Discriminator, which will be discussed further in the section on GAN components.

image.png

Let's break down GANs into three parts:

1. **Generative**: This involves learning a generative model, which explains how data is generated based on a probabilistic model. Essentially, it provides a visual representation of data generation.

2. **Adversarial**: The model is trained in an adversarial setup where two networks compete against each other.

3. **Networks**: GANs use deep neural networks for their training process.

The **generator network** takes random input, often noise, and generates samples like images, text, or audio that resemble the training data. Its goal is to create samples that are indistinguishable from real data.

The **discriminator network** tries to differentiate between real and generated samples. It is trained using both real samples from the training data and generated samples from the generator. Its objective is to correctly identify real data as real and generated data as fake.

The training process involves an adversarial game between the generator and the discriminator. The generator aims to create samples that can deceive the discriminator, while the discriminator strives to enhance its ability to tell real and generated data apart. This adversarial training drives both networks to progressively improve.

As training progresses, the generator becomes increasingly proficient at creating realistic samples, while the discriminator improves its ability to distinguish between real and generated data. Ideally, this process reaches a point where the generator produces high-quality samples that the discriminator finds challenging to differentiate from real data.

GANs have achieved remarkable results in various fields, including image synthesis, text generation, and video generation. They have been employed for tasks such as generating realistic images, creating deepfakes, enhancing low-resolution images, and more. GANs have significantly advanced generative modeling and have unlocked new possibilities for creative applications in artificial intelligence.

### What is Generator?

A Generator in GANs is a neural network designed to create fake data that is used to train the discriminator. It learns to generate convincing data, and the generated samples serve as negative training examples for the discriminator. The Generator takes a fixed-length random vector of noise as input and produces a sample.



#### Generator Network

The main goal of the Generator is to produce outputs that the discriminator will classify as real. The part of the GAN responsible for training the Generator involves the following components:



image.png

- **Noisy Input Vector**: A random vector that serves as the input carrying noise.
- **Generator Network**: Transforms the random input into a data instance.
- **Discriminator Network**: Classifies the generated data as real or fake.
- **Generator Loss**: Penalizes the Generator for failing to deceive the discriminator.



image.png


The backpropagation method is used to adjust each weight in the Generator in the correct direction by calculating the impact of each weight on the output. This process involves obtaining gradients, which are used to update the generator's weights, thereby improving its ability to generate plausible data.


### What is a Discriminator?

The Discriminator in GANs is a neural network that distinguishes real data from the fake data generated by the Generator. It is trained using data from two different sources:

- **Real Data Instances**: These are genuine samples, such as real pictures of birds, humans, currency notes, etc., which serve as positive examples during training.
- **Fake Data Instances**: These are the synthetic samples produced by the Generator, used as negative examples during the training process.

image.png

While training the discriminator, it connects to two loss functions. During discriminator training, the discriminator ignores the generator loss and just uses the discriminator loss.

In the process of training the discriminator, the discriminator classifies both real data and fake data from the generator. The discriminator loss penalizes the discriminator for misclassifying a real data instance as fake or a fake data instance as real.

The discriminator updates its weights through backpropagation from the discriminator loss through the discriminator network.

### How Do GANs Work?

GANs consist of two neural networks: a Generator \( G(x) \) and a Discriminator \( D(x) \). These networks engage in an adversarial game where the Generator's goal is to fool the Discriminator by producing data that closely resembles the training set. The Discriminator's objective is to accurately distinguish between real and fake data. Both networks work simultaneously to learn and process complex data such as audio, video, or image files.

The Generator network takes a random input and generates a synthetic sample. It is trained to maximize the Discriminator's probability of making mistakes, thereby improving its ability to generate realistic data.

Below is an example of how a GAN works to identify whether 100 rupee notes are real or fake. Initially, a noise vector or input vector is fed into the Generator network. The Generator creates fake 100 rupee notes. Both the real images of 100 rupee notes, stored in a database, and the fake notes generated by the Generator are then passed to the Discriminator. The Discriminator classifies the notes as real or fake.

During training, the loss function is calculated at the end of the Discriminator network. This loss is then backpropagated through both the Discriminator and Generator models to update their weights. This process helps the Generator produce more realistic fake notes and improves the Discriminator's ability to distinguish between real and fake notes over time.

### Maths Behind GANs

The math behind Generative Adversarial Networks (GANs) involves a two-player game between a generator and a discriminator, both neural networks. Here's a breakdown of the key concepts:

**Notation:**

* x: Real data from the training dataset.
* z: Random noise vector used as input for the generator.
* G(z): The generated data by the generator, taking a noise vector as input.
* D(x): The discriminator's evaluation of whether the input (x) is real data (probability between 0 and 1).
* D(G(z)): The discriminator's evaluation of the generated data (G(z)), again a probability between 0 and 1.

**The Objective Function:**

GANs are trained using a minimax objective function. This means the generator (G) tries to minimize a loss function, while the discriminator (D) tries to maximize the same function. Here's the core idea:

* The generator wants to create data (G(z)) that fools the discriminator into thinking it's real (D(G(z)) close to 1).
* The discriminator wants to accurately distinguish between real data (D(x)) and fake data (D(G(z))).

**Loss Functions:**

In practice, separate loss functions are often used for the generator and discriminator due to mathematical complexities:

* **Generator Loss:** Minimizes the discriminator's ability to distinguish real from fake data.

  * Common function: log(1 - D(G(z)))
* **Discriminator Loss:** Maximizes its ability to distinguish real from fake data.

  * Common function: log(D(x)) + log(1 - D(G(z)))

### Step for training GANs

* **Define the Problem:** Clearly outline the specific task the GAN will be used for, such as image generation, text generation, or super-resolution.
* **Choose the Architecture of GAN:** Select the appropriate architecture for the Generator and Discriminator networks, considering factors like the type of data and desired output quality.
* **Train Discriminator on Real Data:** Feed real data samples to the Discriminator and update its weights based on its ability to classify them correctly.
* **Generate Fake Inputs for the Generator:** Use random noise vectors as inputs for the Generator to produce fake data samples.
* **Train Discriminator on Fake Data:** Feed the fake data samples generated by the Generator to the Discriminator and update its weights based on its ability to classify them correctly.
* **Train Generator with the Output of the Discriminator:** Update the Generator's weights based on the Discriminator's feedback, aiming to improve the Generator's ability to produce realistic data that fools the Discriminator.

## Different types of GANs


* **Deep Convolutional GANs (DCGANs)**

  The workhorse of image generation! DCGANs leverage convolutional neural networks, particularly adept at processing visual data, for both the generator and discriminator. This makes DCGANs particularly effective in creating high-quality, realistic images.
* **Conditional GANs (CGANs)**

  Unlike vanilla GANs that focus solely on generating realistic data, CGANs incorporate additional information during the generation process. This information, like labels or captions, allows the model to generate data conditioned on that specific input. Imagine a CGAN generating images of cats wearing hats based on a text prompt "cat wearing a hat."
* **Wasserstein GANs (WGANs)**

  Training GANs can be tricky due to issues like vanishing gradients. WGANs address this by using a different objective function during training, leading to more stable training dynamics and potentially better quality outputs.
* **CycleGANs**

  Want to translate an image from one domain to another? CycleGANs are your answer! They involve two GANs working together. One GAN translates data from domain A to domain B, and the other translates data back from B to A. This cycle forces the model to learn a more accurate mapping between the two domains. Imagine using CycleGAN to translate a photo from daytime to nighttime.
* **Variational Autoencoders (VAEs) (Not strictly a GAN)**

  While not technically a GAN, VAEs share some similarities and are often used for generative tasks. VAEs involve an encoder that compresses data into a latent representation and a decoder that reconstructs data from that code. VAEs are useful for tasks like dimensionality reduction and generating new data points that resemble the training data.

```python
def getImage(path,image_size=(64,64)):
    images=[]
    for img in listdir(path):
        temp_image = Image.open(path+'/'+img)
        temp_image = temp_image.convert('RGB')
        image = temp_image.resize(image_size)
        image = np.asarray(image)
        image = ((image - 127.5) / 127.5).astype("float32")
        images.append(image)
    return np.asarray(images)
```

```python
LATENT_DIM = 100
# weight initializer
WEIGHT_INIT = keras.initializers.RandomNormal(mean=0.0, stddev=0.02)
# no. of channels of the image
CHANNELS = 3 # for gray scale, keep it as 1
```

```python
def generator_model():
    model = Sequential(name='generator')

    # 1d random noise
    model.add(layers.Dense(8 * 8 * 512, input_dim=LATENT_DIM))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())

    # convert 1d to 3d
    model.add(layers.Reshape((8, 8, 512)))

    # upsample to 16x16
    model.add(layers.Conv2DTranspose(256, (4, 4), strides=(2, 2), padding='same', kernel_initializer=WEIGHT_INIT))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())

    # upsample to 32x32
    model.add(layers.Conv2DTranspose(128, (4, 4), strides=(2, 2), padding='same', kernel_initializer=WEIGHT_INIT))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())

    # upsample to 64x64
    model.add(layers.Conv2DTranspose(64, (4, 4), strides=(2, 2), padding='same', kernel_initializer=WEIGHT_INIT))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())

    model.add(layers.Conv2D(CHANNELS, (4, 4), padding='same', activation='tanh'))

    return model

generator=generator_model()
generator.summary()
plot_model(generator,show_shapes = True, show_layer_names = True, to_file='Generator_Model.png')
```

```python
def critic_model():
    model = Sequential(name='critic')
    input_shape = (64, 64, 3)
    alpha = 0.2

    # create conv layers
    model.add(layers.Conv2D(64, (4, 4), strides=(2, 2), padding='same', input_shape=input_shape))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU(alpha=alpha))

    model.add(layers.Conv2D(128, (4, 4), strides=(2, 2), padding='same', input_shape=input_shape))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU(alpha=alpha))

    model.add(layers.Conv2D(128, (4, 4), strides=(2, 2), padding='same', input_shape=input_shape))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU(alpha=alpha))

    model.add(layers.Flatten())
    model.add(layers.Dropout(0.3))

    # output class
    model.add(layers.Dense(1, activation='linear'))

    return model

critic=critic_model()
critic.summary()
plot_model(critic,show_shapes = True, show_layer_names = True, to_file='Discriminator_Model.png')
```

```python
def d_wasserstein_loss(pred_real, pred_fake):
    real_loss = tf.reduce_mean(pred_real)
    fake_loss = tf.reduce_mean(pred_fake)
    return fake_loss - real_loss
# Wasserstein loss for the generator
def g_wasserstein_loss(pred_fake):
    return -tf.reduce_mean(pred_fake)
```

```python
class WGAN_GP(keras.Model):
    def __init__(self,
                 critic,
                 generator,
                 latent_dim,
                 critic_extra_steps,
                 gp_weight=10.0): # UPDATE for WGAN-GP: gradient penalty weight
        super().__init__()
        self.critic = critic
        self.generator = generator
        self.latent_dim = latent_dim
        self.c_extra_steps = critic_extra_steps
        self.gp_weight = gp_weight # WGAN-GP
        self.d_loss_metric = keras.metrics.Mean(name="d_loss")
        self.g_loss_metric = keras.metrics.Mean(name="g_loss")


    def compile(self, d_optimizer, g_optimizer, d_loss_fn, g_loss_fn):
        super(WGAN_GP, self).compile()
        self.d_optimizer = d_optimizer
        self.g_optimizer = g_optimizer
        self.d_loss_fn = d_loss_fn
        self.g_loss_fn = g_loss_fn

    @property
    def metrics(self):
        return [self.d_loss_metric, self.g_loss_metric]

    # UPDATE for WGAN-GP: use gradient penalty instead of weight clipping
    def gradient_penalty(self, batch_size, real_images, fake_images):
        """ Calculates the gradient penalty.

        Gradient penalty is calculated on an interpolated image
        and added to the discriminator loss.
        """

        alpha = tf.random.normal([batch_size, 1, 1, 1], 0.0, 1.0)
        diff = fake_images - real_images
        # 1. Create the interpolated image
        interpolated = real_images + alpha * diff

        with tf.GradientTape() as gp_tape:
            gp_tape.watch(interpolated)
            # 2. Get the Critic's output for the interpolated image
            pred = self.critic(interpolated, training=True)

        # 3. Calculate the gradients w.r.t to the interpolated image
        grads = gp_tape.gradient(pred, [interpolated])[0]
        # 4. Calculate the norm of the gradients.
        norm = tf.sqrt(tf.reduce_sum(tf.square(grads), axis=[1, 2, 3]))
        # 5. Calculate gradient penalty
        gradient_penalty = tf.reduce_mean((norm - 1.0) ** 2)
        return gradient_penalty

    def train_step(self, real_images):

        batch_size = tf.shape(real_images)[0]
        noise = tf.random.normal(shape=(batch_size, self.latent_dim))

        # Train the critic more often than the generator by 5 times (self.c_extra_steps)
        for i in range(self.c_extra_steps):
            # Step 1. Train the critic with both real images and fake images
            with tf.GradientTape() as tape:
                pred_real = self.critic(real_images, training=True)
                fake_images = self.generator(noise, training=True)
                pred_fake = self.critic(fake_images, training=True)
                # UPDATE for WGAN-GP: Calculate the gradient penalty
                gp = self.gradient_penalty(batch_size, real_images, fake_images)
                # UPDATE for WGAN-GP: Add gradient penalty to the original critic loss
                d_loss = self.d_loss_fn(pred_real, pred_fake) + gp * self.gp_weight
            # Compute critic gradients
            grads = tape.gradient(d_loss, self.critic.trainable_variables)
            # Update critic weights
            self.d_optimizer.apply_gradients(zip(grads, self.critic.trainable_variables))

        # Step 2. Train the generator (do not update weights of the critic)
        misleading_labels = tf.ones((batch_size, 1)) # G wants D to think the fake images are real (label as 1)

        with tf.GradientTape() as tape:
            fake_images = self.generator(noise, training=True)
            pred_fake = self.critic(fake_images, training=True)
            g_loss = self.g_loss_fn(pred_fake)
        # Compute generator gradients
        grads = tape.gradient(g_loss, self.generator.trainable_variables)
        # Update generator wieghts
        self.g_optimizer.apply_gradients(zip(grads, self.generator.trainable_variables))

        self.d_loss_metric.update_state(d_loss)
        self.g_loss_metric.update_state(g_loss)

        return {"d_loss": self.d_loss_metric.result(), "g_loss": self.g_loss_metric.result()}
```

```python
class GANMonitor(keras.callbacks.Callback):
    def __init__(self, num_img=25, latent_dim=100):
        self.num_img = num_img
        self.latent_dim = latent_dim

        # Create random noise seed for visualization during traing
        self.seed = tf.random.normal([25, latent_dim])

    def on_epoch_end(self, epoch, logs=None):
        generated_images = self.model.generator(self.seed)
        generated_images = (generated_images * 127.5) + 127.5
        generated_images.numpy()

        fig = plt.figure(figsize=(8, 8))
        for i in range(self.num_img):
            plt.subplot(5, 5, i+1)
            img = array_to_img(generated_images[i])
            plt.imshow(img)
            plt.axis('off')
        plt.savefig('epoch_{:03d}.png'.format(epoch))
        plt.show()

    def on_train_end(self, logs=None):
        self.model.generator.save('generator.h5')
```
