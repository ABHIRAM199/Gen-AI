In the modern deep learning era, neural networks excel at a wide range of tasks but typically require large amounts of data to achieve high performance. However, for specific problems like face recognition and signature verification, obtaining extensive datasets is not always feasible. To address this challenge, a new type of neural network architecture called Siamese Networks has been developed.

Siamese Networks can make accurate predictions with only a small number of images. Their ability to learn effectively from minimal data has increased their popularity in recent years. In this article, we will explore what Siamese Networks are and how to develop a signature verification system using PyTorch and Siamese Networks.

## Defination


A Siamese neural network (SNN) is a type of neural network architecture consisting of two or more identical sub-networks. "Identical" means these sub-networks share the same configuration, parameters, and weights. When the parameters are updated in one sub-network, the same updates are applied to the others. This architecture is used to compare feature vectors and identify similarities between inputs.

## Advantages of Siamese Network

* **Effective with Limited Data:** Siamese Networks perform well even when trained on a small dataset. Their ability to learn from minimal data makes them suitable for tasks where acquiring large datasets is challenging.

* **Learning from Pairwise Comparisons:** They are designed to learn from pairwise comparisons, making them particularly useful for tasks like similarity learning, where the emphasis is on comparing input pairs rather than individual inputs.

* **Robust to Variations:** Siamese Networks are inherently robust to variations in input data, such as changes in lighting conditions, pose, or background clutter. This robustness makes them suitable for tasks like face recognition and signature verification, where input variations are common.

* **Embedding Generation:** They produce fixed-length embeddings for each input, which can capture essential features and similarities between inputs. These embeddings can be further used for downstream tasks like clustering or classification.

* **Few-shot Learning:** Siamese Networks excel at few-shot learning, where the model needs to generalize from a small number of examples. This ability makes them suitable for tasks requiring quick adaptation to new classes or concepts.

## Disadvantages of Siamese Network

* **Computational Complexity:** Training Siamese Networks can be computationally expensive, especially when dealing with large datasets or complex architectures. The need to compute pairwise distances or similarities between inputs adds to the computational burden.

* **Sensitive to Hyperparameters:** Siamese Networks often rely on carefully tuned hyperparameters for optimal performance. Choosing the right architecture, learning rate, and margin parameters can significantly impact the network's performance.

* **Training Data Quality:** The effectiveness of Siamese Networks heavily depends on the quality and representativeness of the training data. Poor-quality or biased datasets can lead to suboptimal performance and generalization issues.

* **Limited to Pairwise Comparisons:** While Siamese Networks excel at learning from pairwise comparisons, they may not be well-suited for tasks that require capturing higher-order relationships or dependencies between inputs.

* **Interpretability:** The embeddings generated by Siamese Networks may lack interpretability, making it challenging to understand how the model arrives at its decisions or to diagnose potential issues.

## The Architecture of Siamese Networks

image.png

The Siamese network architecture comprises two identical subnetworks, each handling one of the input data. Initially, the inputs are processed through a convolutional neural network (CNN) to extract essential features from the images. These subnetworks then produce encoded outputs, typically via a fully connected layer, resulting in a condensed representation of the input data.

The CNN consists of two branches and a shared feature extraction component. This shared component includes layers for convolution, batch normalization, ReLU activation, followed by max pooling and dropout layers. The final segment involves the fully connected layer, which maps the extracted features to the ultimate classification outcomes. A function comprises a linear layer followed by a series of ReLU activations and a sequence of operations (convolution, batch normalization, ReLU activation, max pooling, and dropout). The forward function guides the inputs through both branches of the network.

The Differencing layer serves to identify similarities between inputs and enhance differences among dissimilar pairs. This is achieved using the Euclidean Distance function:

Distance(x, x) = f(x)  f(x)

In this context,

- x, x represent the two inputs.
- f(x) represents the output of the encoding.
- Distance denotes the distance function.

This property allows the network to learn effective data representations and apply them to new, unseen samples. As a result, the network generates an encoding, often represented as a similarity score, aiding in class differentiation.

The network operates as a one-shot classifier, eliminating the need for many examples per class.

## Loss functions used in Siamese Networks

A loss function serves as a mathematical metric to quantify the disparity between the expected and observed output within a machine learning model, considering a particular input. During the model training process, the objective is to minimize this loss function by fine-tuning the model's parameters.

Various loss functions are tailored to address different types of problems. For example, mean squared error is suitable for regression problems, whereas cross-entropy loss is more fitting for classification tasks.

Unlike several other types of neural networks, the Siamese Network adopts multiple loss functions, which are detailed below.

### Binary Cross-Entropy Loss

Siamese networks can be trained to tell if two things are similar, like identifying matching faces in photos. To achieve this, the network predicts a score between 0 and 1, indicating how likely it thinks the two things are similar. A score close to 1 means very similar, and a score close to 0 means very different.

We train the network by showing it pairs of images, labeled as "similar" or "different." The network's goal is to get its predicted similarity scores to match these labels as closely as possible. To measure how well it's doing, we use a special function called binary cross-entropy loss. This function basically calculates how much the network's predictions differ from the correct labels.

The lower the loss, the better the network is at distinguishing similar and different images. By adjusting the network's internal settings based on this loss, we train it to become a better similarity detector.

### Contrastive Loss

Siamese networks use a special trick called contrastive loss to learn how similar images are. This trick works by measuring the distance between the codes the network creates for each image. Smaller distances indicate more similar images, while larger distances mean the images are different.

This approach is particularly helpful when you don't have many training examples. Contrastive loss only needs you to feed the network pairs of images, labeled as either "similar" (positive pairs) or "different" (negative pairs). The network then learns to push the codes for similar images closer together and dissimilar images further apart.

The contrastive loss function is like a formula that helps the Siamese network learn how similar images are. It considers two things: the distance between the codes the network creates for each image (called Euclidean distance) and whether the images are supposed to be similar or different (represented by the input parameter Y).

If the images are supposed to be similar (Y = 0): The function only cares about how close their codes are (D^2). The smaller the distance, the better (because similar images should have similar codes).
If the images are supposed to be different (Y = 1): The function only cares if the distance between their codes is larger than a certain margin (m). The bigger the gap between the codes (as long as it's bigger than m), the better (because dissimilar images should have very different codes).
This margin (m) is a helpful tuning knob. It prevents the network from getting fooled by images that are slightly similar but still different. The margin essentially sets a minimum distance for images to be considered truly dissimilar.

### Triplet Loss

The triplet loss uses triples of data. The graphic below illustrates these triples.

image.png

The triplet loss function aims to enhance the separation between the anchor and negative samples while reducing the gap between the anchor and positive samples.

Mathematically, the Triplet loss function defines itself as the maximum difference between the anchor-to-positive distance (d(a,p)) and the anchor-to-negative distance (d(a,n)), subtracted by a margin value. When this difference is positive, the computed value becomes the loss; otherwise, it is set to zero.

Heres a breakdown of the components:

* d signifies the Euclidean distance.
* a represents the anchor input.
* p denotes the positive input.
* n stands for the negative input.

The primary goal is to ensure that the positive input is closer to the anchor input than the negative input, maintaining a margin of separation.

## Applications of Siamese Networks

1. **Face Recognition:**
Siamese networks can be used for face recognition tasks by learning embeddings of faces in a common feature space. Given a pair of face images, the network can determine if they belong to the same person or not.

2. **Signature Verification:**
In signature verification systems, Siamese networks compare pairs of signatures to authenticate individuals. They learn to extract discriminative features from signatures, enabling accurate verification.

3. **Similarity Search:**
Siamese networks excel in similarity search tasks such as image retrieval and product recommendation. By learning embeddings of images or items, the network can efficiently retrieve similar items from a large database.

4. **Text Similarity:**
For natural language processing tasks, Siamese networks can measure the semantic similarity between texts. They learn to encode text inputs into fixed-size vectors, enabling tasks like duplicate detection and paraphrase identification.

5. **Medical Diagnosis:**
Siamese networks find applications in medical diagnosis tasks such as disease classification and medical image analysis. They can learn to compare medical images or patient data to assist in diagnosing diseases or conditions.

6. **Anomaly Detection:**
Siamese networks are used for anomaly detection in various domains, including cybersecurity and industrial machinery. They learn to distinguish between normal and anomalous behavior by comparing input patterns.

7. **Speaker Verification:**
In speaker verification systems, Siamese networks compare pairs of speech samples to determine if they belong to the same speaker. They learn to extract speaker embeddings, facilitating accurate speaker verification.

8. **Document Matching:**
Siamese networks are employed in document matching tasks such as plagiarism detection and document clustering. They learn to compare pairs of documents and assess their similarity or dissimilarity.

9. **Gesture Recognition:**
In gesture recognition applications, Siamese networks compare pairs of gesture sequences to recognize specific gestures or actions. They learn to extract temporal features from gesture data for accurate recognition.

10. **Object Tracking:**
Siamese networks can be utilized for object tracking in videos by comparing the appearance of objects across frames. They learn to track objects by estimating the similarity between object representations.

### Defining data loader class

```python
class DataLoader:
  #constructor
  def __init__(self, dataset, batch_size,dir):
    self.dataset = dataset
    self.batch_size = batch_size
    self.dir = dir
  #shuffler
  def shuffle(self):
    return self.dataset.sample(frac=1)
  #generator
  def datagen(self):
    num_samples = len(self.dataset)
    while True:
        # shuffling the samples
        self.dataset = self.shuffle()
        for batch in range(1, num_samples, self.batch_size):
            image1_batch_samples = self.dir + "/" + self.dataset.iloc[:, 0][batch:batch + self.batch_size]
            image2_batch_samples = self.dir + "/" + self.dataset.iloc[:, 1][batch:batch + self.batch_size]
            label_batch_samples = self.dataset.iloc[:, 2][batch:batch + self.batch_size]
            Image1, Image2, Label = [], [], []
            for image1, image2, label in zip(image1_batch_samples, image2_batch_samples, label_batch_samples):
                # append them to Images directly
                image1_data = Image.open(image1)
                image2_data = Image.open(image2)
                # resizing the images
                image1_data = image1_data.resize((112, 112))
                image2_data = image2_data.resize((112, 112))
                # converting to array
                image1_data = img_to_array(image1_data)
                image2_data = img_to_array(image2_data)
                Image1.append(image1_data)
                Image2.append(image2_data)
                Label.append(label)
            # convert each list to numpy arrays to ensure that they get processed by fit function
            Image1 = np.asarray(Image1).astype(np.float32)
            Image2 = np.asarray(Image2).astype(np.float32)
            Label = np.asarray(Label).astype(np.float32)
            yield [Image1, Image2], Label
```

### Splitting the data into train & val set

```python
train_set, val_set = train_test_split(train_dataset, test_size=0.25)
```

###  Creating Dataloaders

```python
train_gen= DataLoader(train_set,1024,train_dir)
val_gen = DataLoader(val_set,1024,train_dir)
```

### Performing sanity check of the dataloader

```python
train_batch = next(train_gen.datagen())
print("Train batch images shape:", train_batch[0][0].shape, train_batch[0][1].shape)
print("Train batch labels shape:", train_batch[1].shape)
```

### Defining custom siamese model architecture

```python
def custom_siamese_model(input_shape):
    input1 = Input(input_shape)
    input2 = Input(input_shape)

    model = Sequential()
    model.add(Conv2D(16, (3,3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D(2,2))
    model.add(Dropout(0.25))

    model.add(Conv2D(32, (3,3), activation='relu'))
    model.add(MaxPooling2D(2,2))
    model.add(Dropout(0.25))

    model.add(Conv2D(64, (3,3), activation='relu'))
    model.add(MaxPooling2D(2,2))
    model.add(Dropout(0.25))

    model.add(Conv2D(128, (3,3), activation='relu'))
    model.add(MaxPooling2D(2,2))
    model.add(Dropout(0.25))

    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dense(128, activation='relu'))

    # Call the model with the inputs
    embedding1 = model(input1)
    embedding2 = model(input2)

    loss_layer = Lambda(lambda tensors: K.abs(tensors[0] - tensors[1]))
    manhattan_distance = loss_layer([embedding1, embedding2])
    output = Dense(1, activation='sigmoid')(manhattan_distance)
    network = Model(inputs=[input1, input2], outputs=output)
    return network
```

### Instantiating early stopper with custom callbacks

```python
early_stopper =  EarlyStopping(monitor='val_loss',min_delta=0,patience=3,verbose=1)
custom_callback = [early_stopper]
```

### Instantiating model

```python
model = custom_siamese_model((112,112,3))
model.summary()
```

### Defining the model and compiling it

```python
optimizer = Adam(lr = 1e-4)
model.compile(loss="binary_crossentropy",optimizer=optimizer,metrics=['accuracy'])
```

### Training the model

```python
print("Initializing Training !!")
history = model.fit_generator(
    train_gen.datagen(),
    verbose=1,
    steps_per_epoch=100,  # set appropriate steps_per_epoch
    epochs=5,
    validation_data=val_gen.datagen(),
    validation_steps=100,  # set appropriate validation_steps
    callbacks=custom_callback
)
```

###  Plotting the accuracy and loss plots

```python
history_keys = history.history.keys()

plt.figure(figsize=(12, 4))

# Plot the accuracy curves
if 'accuracy' in history_keys and 'val_accuracy' in history_keys:
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'], loc='lower right')

# Plot the loss curves
if 'loss' in history_keys and 'val_loss' in history_keys:
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'], loc='upper right')

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
```
