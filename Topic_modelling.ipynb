# Introduction to Parts of Speech (POS) Tagging



## Parts of Speech (POS) Tagging

image.png

Part-of-Speech (POS) tagging is a technique in natural language processing that involves assigning grammatical categories, such as nouns, verbs, adjectives, adverbs, and pronouns, to each word in a sentence. This process helps to reveal the syntactic structure of the text, which is essential for understanding the relationships between words, disambiguating word meanings, and enabling various linguistic and computational analyses.

During our school days, we learned about parts of speech, including nouns, pronouns, adjectives, and verbs. Words from different parts of speech come together to form sentences. Recognizing the part of speech of each word in a sentence is crucial for comprehending its meaning.

This necessity led to the development of POS tagging. As you might have already inferred, POS tagging is the method of assigning specific labels, known as POS tags, to each word in a sentence to indicate its part of speech.

### Universal Part-of-Speech (POS)

Universal Part-of-Speech (POS) tags provide a standardized set of categories to label words across different languages. This standardization is crucial for creating multilingual NLP applications and for comparative linguistic studies. The Universal POS tags are designed to be simple and generalizable, covering the basic grammatical categories found in many languages. Here is a list of the Universal POS tags along with brief descriptions and examples:

* **ADJ (Adjective)**

  * Modifies a noun by describing a quality or attribute.
  * Example: "happy", "blue", "large"
* **ADP (Adposition)**

   * Includes prepositions and postpositions, indicating relationships between different elements within a sentence.
   * Example: "in", "at", "on"
* **ADV (Adverb)**

  * Modifies a verb, an adjective, or another adverb, often indicating manner, place, time, or degree.
  * Example: "quickly", "very", "there"
* **AUX (Auxiliary)**

   * Supports the main verb in a sentence by providing additional grammatical information such as tense, mood, or voice.
   * Example: "is", "have", "will"
* **CCONJ (Coordinating Conjunction)**

  * Connects words, phrases, or clauses of equal syntactic importance.
  * Example: "and", "but", "or"
* **DET (Determiner)**

  * Introduces nouns and specifies their reference within the context.
  * Example: "the", "a", "this"
* **INTJ (Interjection)**

  * Expresses an emotion or a spontaneous reaction.
  * Example: "oh", "wow", "ouch"
* **NOUN (Noun)**

  * Refers to a person, place, thing, or idea.
  * Example: "dog", "city", "happiness"
* **NUM (Numeral)**

  * Expresses a number or a relation to a number.
  * Example: "one", "two", "second"
* **PART (Particle)**

  * Includes function words that do not fit neatly into the standard part-of-speech categories.
  * Example: "not", "to" (as in "to go")
* **PRON (Pronoun)**

  * Replaces a noun or a noun phrase to avoid repetition.
  * Example: "he", "she", "it"
* **PROPN (Proper Noun)**

  * Refers to specific names of people, places, organizations, etc.
  * Example: "John", "London", "UNESCO"
* **PUNCT (Punctuation)**

  * Marks the boundaries of sentences, phrases, or clauses.
  * Example: ".", ",", "?"
* **SCONJ (Subordinating Conjunction)**

  * Introduces a subordinate clause and shows the relationship between the subordinate clause and the main clause.
  * Example: "because", "although", "if"
* **SYM (Symbol)**

  * Represents various symbols and signs not included in the usual alphanumeric set.
  * Example: "$", "%", "&"
* **VERB (Verb)**

  * Describes an action, event, or state of being.
  * Example: "run", "eat", "exist"

## Advantages of POS Tagging

**Improved Text Understanding:**

* By identifying the grammatical role of each word (noun, verb, adjective, etc.), POS tagging helps computers grasp the basic structure and meaning of a sentence.
This understanding is crucial for various NLP tasks that rely on analyzing sentence components and relationships between words.

**Disambiguation of Words:**

* Many words have multiple meanings depending on their context (e.g., "play" as a noun or verb). POS tagging helps resolve this ambiguity by identifying the most likely part of speech based on the surrounding words. This clarifies the intended meaning and improves overall comprehension.


**Information Extraction:**

* Extracting specific information from text relies on identifying relevant entities and their relationships, which can be facilitated by POS tagging.
Improved Information Retrieval:

* Search engines and information retrieval systems can leverage POS tagging to refine search results. By understanding the parts of speech in a query, the system can better match relevant documents and improve search accuracy.

**Simpler Text Analysis:**

* POS tagging offers a way to analyze large amounts of text efficiently. By categorizing words grammatically, it allows for identifying patterns and trends in word usage across a corpus. This can be valuable for tasks like sentiment analysis or topic modeling.

## Disadvantages of POS Tagging

**Ambiguity:**

* Many words can have multiple POS tags depending on the context. For instance, "play" can be a noun (a children's play) or a verb (to play a game). POS taggers might struggle to choose the correct tag in these cases, especially for uncommon usages.

**Out-of-Vocabulary (OOV) Words:**

* POS taggers are trained on a massive dataset of words and their tags. However, they can't perfectly predict the tags for words they haven't encountered during training (OOV words). This can be problematic for new words, slang, or technical terms.

**Domain Dependence:**

* POS taggers trained on general text might not perform well on specific domains like legal documents or medical reports. The language used in these domains has unique characteristics, and a tagger trained on generic data might struggle to handle them accurately.

**Accuracy Limitations:**

* No POS tagger is perfect. Even with large training datasets, errors can occur due to complex sentence structures, sarcasm, or unusual phrasings. The accuracy of a POS tagger depends on the specific model and the quality of the training data.

**Computational Cost:**

* Training complex POS taggers, especially deep learning models, can be computationally expensive. This requires significant resources and processing power, which might not be feasible for everyone.

**Limited Insight:**

* POS tagging only provides information about the grammatical function of words, not their meaning. This can be limiting for tasks that require deeper semantic understanding, like sentiment analysis or topic modeling.

## POS Tagging Approaches

1. **Rule-Based POS Tagging**

   Rule-Based POS Tagging is an approach to assigning Part-of-Speech (POS) tags to words in a sentence based on a set of handcrafted linguistic rules. Here's a more detailed overview of this approach:


   * Linguistic Rules: Rule-Based POS Tagging relies on linguistic rules that encode patterns, heuristics, and syntactic properties of natural language.
   * Pattern Matching: These rules typically involve matching words or sequences of words against predefined patterns or criteria to determine their POS tags.
   * Lexicons and Dictionaries: Rule-Based taggers may also utilize lexicons or dictionaries to look up known words and their associated POS tags.
   * Hierarchical Rules: Rules may be organized hierarchically to handle complex linguistic phenomena, such as compound words, idiomatic expressions, or syntactic structures.
   
     **Example**

   Here's a simplified example of a rule for assigning POS tags to determiners in English:

   If a word is "the", it is tagged as a determiner (DT).
   If a word is "a" or "an", it is tagged as a determiner (DT).

  **Advantages**

   * Interpretability: Rule-Based taggers are transparent and easy to interpret because they explicitly encode linguistic rules.

   * Domain-Specific Rules: Linguists or domain experts can customize the rules to handle specific linguistic phenomena or domain-specific vocabulary.

   * Language-Independent: Rule-Based taggers can be adapted to different languages by crafting rules based on the grammatical structures and patterns of each language.

  **Disadvantages**

  * Limited Coverage: Rule-Based taggers may struggle with handling irregularities, exceptions, or unseen linguistic phenomena not covered by the rules.
  * Manual Effort: Creating and maintaining linguistic rules requires domain expertise and manual effort, making it labor-intensive and time-consuming.
  * Generalization: Rule-Based taggers may not generalize well across different domains or languages without extensive rule adaptation and tuning.

  **Applications:**

  * Rule-Based POS Tagging can be useful in scenarios where domain-specific linguistic knowledge is crucial, such as specialized domains like legal or medical text processing.

  * It can also serve as a baseline or complementary approach to more data-driven methods, providing human-readable insights into the tagging process.


2. **Probabilistic POS Tagging**

   Probabilistic POS Tagging is an approach to assigning Part-of-Speech (POS) tags to words in a sentence based on statistical models and probabilities derived from annotated corpora. Here's a detailed overview of this approach:

   **Description:**

   * Statistical Models: Probabilistic POS Tagging relies on statistical models trained on annotated corpora, where each word in a sentence is associated with a probability distribution over possible POS tags.
   * Hidden Markov Models (HMMs): One common probabilistic model used for POS tagging is the Hidden Markov Model, which models the joint probability of observing a sequence of words and their corresponding POS tags.
   * Maximum Entropy Models (MaxEnt): Another popular approach is Maximum Entropy Markov Models, which learn to predict the most likely POS tag for a word given its context (neighboring words) and linguistic features.
   * N-gram Models: Some probabilistic taggers use N-gram models to capture the contextual dependencies between words and POS tags within a fixed-size window of preceding words.

   **Training:**

   * Annotated Corpora: Probabilistic taggers are trained on large annotated corpora, such as the Penn Treebank, which provide labeled examples of word sequences and their corresponding POS tags.
   * Feature Extraction: During training, the tagger extracts linguistic features (e.g., word identity, suffixes, prefixes, context words) from the training data to inform the tagging decisions.
   * Parameter Estimation: The tagger learns the parameters of the statistical model, such as transition probabilities between POS tags or emission probabilities of words given POS tags, using optimization algorithms like maximum likelihood estimation or gradient descent.

   **Inference:**

   * Decoding Algorithms: During inference, the tagger uses decoding algorithms such as the Viterbi algorithm to find the most likely sequence of POS tags given the observed words and the trained model.
   * Smoothing Techniques: To handle unseen words or rare linguistic phenomena, probabilistic taggers often employ smoothing techniques to adjust the probability distributions and prevent zero probabilities.

   **Advantages**
   * Data-Driven: Probabilistic taggers learn from data and can capture complex linguistic patterns and dependencies present in annotated corpora.
   * Language-Independent: They can be applied to different languages by training on appropriately annotated data, making them adaptable and scalable.
   * Domain Adaptation: Probabilistic taggers can be fine-tuned or adapted to specific domains or genres by retraining on domain-specific corpora.

   **Disadvantages**

   * Data Requirements: Probabilistic taggers require large annotated corpora for training, which may not be available for all languages or domains.
   * Overfitting: They may overfit to the training data, leading to poor generalization performance on unseen or out-of-domain text.
   * Complexity: Some probabilistic models, such as HMMs, have computational complexity issues, especially for long sequences of words.

   **Applications:**

   * Probabilistic POS Tagging is widely used in various NLP applications, including text processing, information retrieval, machine translation, and speech recognition.
   * It serves as a foundational component in many downstream NLP tasks, such as syntactic parsing, named entity recognition, and sentiment analysis.

### Other POS Tagging Approaches

3. **Rule-Based + Probabilistic Hybrid Approaches:**
   * Description: These approaches combine rule-based methods with probabilistic models to leverage the strengths of both approaches.
   * Pros: They can achieve higher accuracy by incorporating linguistic knowledge from rules and statistical patterns from data. They're flexible and can be tailored to specific tasks or domains.
   * Cons: They require expertise in both rule-based and statistical methods for development and maintenance.
4. **Neural Network-Based POS Tagging:**
  * Description: This approach utilizes neural network architectures, such as Recurrent Neural Networks (RNNs), Convolutional Neural Networks (CNNs), or Transformer models, to learn POS tagging directly from raw text.
  * Pros: It can automatically learn complex patterns and representations from data without relying on handcrafted features or rules. It can achieve state-of-the-art performance on large datasets.
  * Cons: It requires large amounts of annotated data for training and significant computational resources for training deep learning models. It may lack interpretability compared to rule-based or probabilistic approaches.
5. **Transfer Learning-Based POS Tagging:**
  
   * Description: This approach fine-tunes pre-trained language models, such as BERT, GPT, or ELMo, on POS tagging tasks to leverage pre-existing linguistic knowledge and representations.
   
   * Pros: It can achieve high performance with limited annotated data by leveraging pre-trained language representations. It can handle various languages and domains with minimal task-specific tuning.
   
   * Cons: It requires large pre-trained models and computational resources for fine-tuning. It may not perform optimally for languages or domains significantly different from the pre-training data.
6. **Ensemble Approaches:**
   * Description: Ensemble methods combine predictions from multiple POS tagging models, such as rule-based, probabilistic, neural network-based, or transfer learning-based models, to improve overall performance.
   
   * Pros: They can mitigate individual model biases and errors, leading to more robust and accurate predictions. They can leverage diverse sources of information and expertise.
   
   * Cons: They can be computationally expensive and complex to implement. They require careful selection and combination of diverse models to achieve optimal performance.
7. **Linguistically-Motivated POS Tagging:**
   * Description: This approach incorporates linguistic theories and insights to design POS tagging models and features, such as morphological analysis, syntactic structure, or semantic information.
   * Pros: It can capture rich linguistic phenomena and improve generalization across languages and domains. It can provide interpretable and linguistically meaningful results.
   * Cons: It may require domain-specific linguistic expertise and annotation efforts. It may be less scalable or adaptable to new languages or domains compared to data-driven approaches.

## Dependency parsing

Dependency parsing is a process in natural language processing (NLP) that involves analyzing the grammatical structure of a sentence to establish relationships between "head" words and words that modify those heads. This analysis helps to understand the syntactic structure and meaning of a sentence.

* **Key Concepts**
  * Head: A central word in a sentence that determines the syntactic type of the phrase. For example, in the phrase "the big cat," "cat" is the head.

  * Dependent: A word that modifies or complements the head. In the phrase "the big cat," "the" and "big" are dependents of "cat."

  * Dependency Relation: The type of syntactic relationship between a head and its dependent. Common relations include nsubj (nominal subject), dobj (direct object), and amod (adjectival modifier).

**Example**

Consider the sentence: "She reads a book."

"She" is the subject (nsubj) of the verb "reads."

"reads" is the main verb (root) of the sentence.

"a" is a determiner (det) modifying "book."

"book" is the direct object (dobj) of the verb "reads."


### Dependency Parsing with spaCy

```python
import spacy

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Define the text to be analyzed
text = 'She reads a book.'

# Process the text with the spaCy model
doc = nlp(text)

# Print the dependency relations for each token
print(f"{'Token':<10} {'Head':<10} {'Dep. Relation':<15} {'POS':<10}")
print("="*50)
for token in doc:
    print(f"{token.text:<10} {token.head.text:<10} {token.dep_:<15} {token.pos_:<10}")

# Visualize the dependency parse
spacy.displacy.render(doc, style='dep', jupyter=True)
```

## Constituency parsing

Constituency parsing, also known as phrase structure parsing, is a natural language processing (NLP) technique that involves analyzing the grammatical structure of a sentence by breaking it down into its constituent phrases. These phrases are organized hierarchically according to the rules of a formal grammar, typically a context-free grammar (CFG). Constituency parsing aims to identify the syntactic relationships between words and group them into phrases such as noun phrases, verb phrases, prepositional phrases, etc.

**Key Concepts**
* Constituent: A group of words that function as a unit within a sentence, forming a syntactic unit. Examples include noun phrases (NP), verb phrases (VP), prepositional phrases (PP), etc.

* Parse Tree: A hierarchical representation of the syntactic structure of a sentence, where each node represents a constituent and its children represent its immediate sub-constituents.

** Constituency Parsing Methods**

* Rule-based Parsing: Utilizes predefined grammar rules to generate parse trees. This approach often involves manually crafted grammar rules.

* Statistical Parsing: Learns parsing patterns from annotated corpora using statistical models such as probabilistic context-free grammars (PCFGs), maximum entropy models, or neural networks.

**Example**
Consider the sentence: "The cat chased the mouse under the table."

A constituency parse tree for this sentence might look like this:

image.png

In this parse tree:

* The top-level node represents the entire sentence.
* The sentence is divided into two main constituents: a noun phrase (NP) and a verb phrase (VP).
* Each constituent is further divided into smaller constituents until individual words are reached.

### Constituency Parsing with NLTK

```python
import nltk

# Define a simple context-free grammar
cfg_grammar = nltk.CFG.fromstring("""
    S -> NP VP '.'
    NP -> DT N | DT N PP
    VP -> V NP | V NP PP
    PP -> P NP
    DT -> 'The' | 'the'
    N -> 'cat' | 'mouse' | 'table'
    V -> 'chased'
    P -> 'under'
""")

# Create a parser using the custom CFG grammar
parser = nltk.ChartParser(cfg_grammar)

# Define the sentence
sentence = "The cat chased the mouse under the table."

# Tokenize the sentence
tokens = nltk.word_tokenize(sentence)

# Perform constituency parsing using the custom grammar
for tree in parser.parse(tokens):
    # Print the parse tree in ASCII format
    print(tree)
```

* (S ...) represents the sentence structure,
* (NP ...) represents the noun phrase,
* (VP ...) represents the verb phrase,
* (DT ...) represents the determiner,
* (N ...) represents the noun,
* (V ...) represents the verb,
* (PP ...) represents the prepositional phrase, and
* (P ...) represents the preposition.

* Syntactic Analysis: Understanding the grammatical structure of sentences.

* Question Answering: Analyzing sentence structure to extract relevant information.
* Text-to-Speech Conversion: Generating natural-sounding speech by understanding sentence structure.

## Topic Modeling

Topic modeling stands as a pivotal tool in statistical analysis and natural language processing, designed to unveil the underlying “topics” inherent in a collection of documents. It represents a potent technique in text mining, facilitating the uncovering of implicit semantic structures within textual data. Through the application of topic modeling, scholars and analysts can glean insights into the fundamental themes and concepts encapsulated within the texts under scrutiny.

Diverging from rule-based text mining methods reliant on regular expressions or dictionary-based keyword searches, topic modeling operates as an unsupervised approach aimed at identifying and scrutinizing clusters of words, termed “topics,” within extensive bodies of text.

Topics can be characterized as “recurring patterns of co-occurring terms within a corpus.” A proficient topic model should yield clusters such as “health,” “doctor,” “patient,” and “hospital” for a topic centered around healthcare, or “farm,” “crops,” and “wheat” for a topic focused on farming.

Topic models serve manifold purposes, including document clustering, the organization of vast textual datasets, information retrieval from unstructured text, and feature selection. For instance, organizations like The New York Times leverage topic models to enhance their user-article recommendation engines. Additionally, professionals in the recruitment sector employ topic models to extract latent features from job descriptions and align them with suitable candidates. They are also utilized for organizing extensive collections of emails, customer reviews, and user profiles across social media platforms.

### Latent Dirichlet Allocation for Topic Modeling

Numerous methods exist for extracting topics from text, including Term Frequency-Inverse Document Frequency and Non-Negative Matrix Factorization techniques. However, Latent Dirichlet Allocation (LDA) stands out as the most widely used topic modeling technique, which we'll delve into in this article.

LDA operates on the assumption that documents are generated from a blend of various topics, with each topic generating words according to its probability distribution. When presented with a dataset of documents, LDA retraces its steps to infer what topics could have generated those documents initially.

At its core, LDA is a matrix factorization technique. In the vector space, any corpus (a collection of documents) can be depicted as a document-term matrix. This matrix represents a corpus consisting of N documents (D1, D2, D3, ..., Dn) and a vocabulary size of M words (W1, W2, ..., Wn). The value in the i,j cell of the matrix signifies the frequency count of word Wj in Document Di.

image.png

LDA transforms the Document-Term Matrix into two lower-dimensional matrices: M1 and M2. M1 represents a document-topics matrix, while M2 represents a topic-terms matrix. These matrices have dimensions of (N, K) and (K, M) respectively, where N is the number of documents, K is the number of topics, and M is the vocabulary size.

image.png



Initially, these matrices provide distributions for topic words and document topics. However, LDA aims to refine these distributions further. It achieves this by employing sampling techniques.

The process involves iterating through each word “w” in each document “d” and adjusting the current topic-word assignment with a new one. A new topic “k” is assigned to word “w” with a probability P, which is the product of two probabilities: p1 and p2.

For each topic, two probabilities, p1 and p2, are computed. P1 represents the proportion of words in document d that are currently assigned to topic t, while P2 represents the proportion of assignments to topic t over all documents that originate from this word w.

The current topic-word assignment is then updated with a new topic based on the probability, the product of p1 and p2. During this step, the model assumes that all existing word-topic assignments except the current word are correct. This adjustment reflects the probability that topic t generated word w, thus rationalizing the update.

After multiple iterations, the model reaches a steady state where the distributions for document topics and topic terms become fairly accurate. This convergence point marks the completion of LDA's process.

### Parameters of LDA


The Alpha and Beta hyperparameters play crucial roles in the Latent Dirichlet Allocation (LDA) model. Alpha represents the document-topic density, while Beta represents the topic-word density.

* **Alpha Hyperparameter:** A higher value of alpha indicates that documents are composed of more topics, while a lower value suggests that documents contain fewer topics. Alpha essentially controls the distribution of topics in documents.

* **Beta Hyperparameter:** On the other hand, a higher value of beta means that topics are composed of a large number of words in the corpus. Conversely, a lower value of beta implies that topics consist of fewer words. Beta regulates the distribution of words in topics.

The choice of these hyperparameters can significantly influence the outcome of the topic modeling process. Researchers often experiment with different values of alpha and beta to find the configuration that best fits their data and objectives.

The Number of Topics is another critical parameter in LDA. This determines the number of topics to be extracted from the corpus. While there is no definitive rule for selecting the number of topics, researchers have developed approaches, such as using the Kullback-Leibler Divergence Score, to obtain an optimal number of topics. This score helps in evaluating the quality of the topics generated by the model. However, delving into the mathematical details of this score is beyond the scope of this discussion.

The Number of Topic Terms refers to the number of terms composed in a single topic. This parameter is typically decided based on the specific requirements of the problem. For tasks involving extracting themes or concepts, a higher number of topic terms may be preferred. Conversely, if the goal is to extract specific features or terms, a lower number may be more suitable.

### Exploring Topic Modeling Techniques

Your comparison between Latent Semantic Analysis (LSA) and Latent Dirichlet Allocation (LDA) is quite comprehensive and accurate. Let's summarize the key points:

* **Latent Semantic Analysis (LSA):**
 * Methodology: Utilizes Singular Value Decomposition (SVD) to reduce the dimensionality of the term-document matrix.
 * Assumption: Assumes that words with similar meanings will appear in similar documents.
 * Objective: Seeks to capture the underlying relationships between words to determine document similarity.
 * Representation: Represents documents and words in a lower-dimensional space where similarities between documents are preserved.
 * Application: Widely used in information retrieval tasks and search engines.
* **Latent Dirichlet Allocation (LDA):**
 * Methodology: Bayesian generative model that assigns topics to documents and words probabilistically.
 * Assumption: Assumes that each document is a mixture of a small number of latent topics, and each word is generated by a particular topic.
 * Objective: Aims to discover the latent topics present in a corpus of text and their distribution across documents.
 * Representation: Represents documents as distributions over topics and topics as distributions over words.
 * Application: Commonly used in document clustering, topic modeling, and content recommendation systems.

* **Comparison:**
 * Underlying Assumptions: LSA assumes word co-occurrence patterns reflect semantic similarity, while LDA assumes documents are generated by a mixture of latent topics.
 * Methodology: LSA uses matrix factorization techniques, while LDA employs probabilistic modeling.
 * Focus: LSA focuses on capturing word-document relationships, while LDA focuses on identifying underlying topics.
 * Interpretability: LDA provides more interpretable results in terms of identifying topics, while LSA is more focused on document similarity.

### Extract Noun Phrases

```python
def get_noun_phrases(text):
    doc = nlp(text)
    required_nc = []  # Initialize an empty list to store the filtered noun phrases
    for chunk in doc.noun_chunks:
        new_nc = [token.text for token in chunk if token.pos_ in ["NOUN", "PROPN"]]
        if new_nc:
            new_nc_text = " ".join(new_nc)  # Join the tokens to form a single noun phrase
            required_nc.append(new_nc_text)  # Append the filtered noun phrase to the list
    return required_nc
```

### Improve results of topic modeling

Absolutely, the success of topic modeling hinges greatly on the features, or terms, present in the corpus. Since the corpus is typically represented as a document-term matrix, which tends to be sparse, reducing its dimensionality can significantly enhance the effectiveness of topic modeling. Here are a few practical approaches based on my experience:

**1. Frequency Filter:**
* **Sorting by Frequency:** Arrange terms based on their frequency of occurrence in the corpus. Terms with higher frequencies are more likely to contribute meaningfully to the topics compared to those with lower frequencies.
* **Threshold Selection:** Conduct an exploratory analysis of terms and their frequencies to determine a suitable threshold. Terms below this threshold frequency can be considered weak features and can be excluded from further analysis.
2. Part of Speech Tag Filter:**
* **Contextual Relevance**: Part-of-speech (POS) tag filtering focuses on the contextual importance of features rather than their frequencies alone.
* **POS Tag Analysis**: Certain POS tags, such as prepositions (IN), cardinal numbers (CD), and modal verbs (MD), often contain terms that are less semantically significant in topic modeling.
* **Filtering Mechanism**: By studying the POS tags of terms in the corpus, we can identify and filter out terms that are deemed less important contextually. For example, terms like "within," "upon," "except" (IN), "one," "two," "hundred" (CD), and "may," "must" (MD) may be removed from consideration.

Applying a POS tag filter helps in focusing the topic modeling process on terms that carry more semantic weight and relevance, thereby improving the quality and interpretability of the resulting topics.


**3. Batch-Wise LDA:**
   - **Enhanced Topic Term Retrieval**: Batch-wise LDA involves dividing the corpus into batches of fixed sizes and running the LDA algorithm separately on each batch.
   - **Multiple Iterations**: Running LDA multiple times on different batches yields varying results due to the stochastic nature of the algorithm.
   - **Identifying Important Topic Terms**: Despite the variation in results, the most important topic terms are those that consistently appear across all batches.
   - **Intersection of Results**: By taking the intersection of the topic terms extracted from each batch, we can retrieve the most robust and representative set of terms for each topic.

Batch-wise LDA enhances the retrieval of important topic terms by leveraging multiple iterations and consolidating the results across different subsets of the corpus. This approach improves the stability and reliability of the topic modeling process.
